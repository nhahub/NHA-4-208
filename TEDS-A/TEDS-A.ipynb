```python
import polars as pl
df2021=pl.read_csv("D:\\DEPI\\tedsa_puf_2021.csv")
df2022=pl.read_csv("D:\\DEPI\\tedsa_puf_2022.csv")
df2023=pl.read_csv("D:\\DEPI\\tedsa_puf_2023.csv")
```

```python
all_data = pl.concat(
    [df2021, df2022, df2023],
    how="vertical"
)

all_data.write_parquet(
    "tedsa_2021_2023.parquet"
)
```

```python
all_data = all_data.with_columns(
    pl.all().replace(-9, None)
)
```

```python
all_data.null_count()
```

```python
all_data = all_data.drop_nulls(
    subset=["SUB1"]
)

all_data.null_count()
```

```python
important_columns = [
    "ADMYR",
    "AGE",
    "SEX",
    "SUB1",
    "ROUTE1",
    "FREQ1",
    "FRSTUSE1",
    "SERVICES",
    "PSYPROB",
    "STFIPS",
    "RACE",
    "ETHNIC",
    "EDUC",
    "EMPLOY",
    "MARSTAT",
    "PSOURCE",
    "PRIMPAY",
    "HLTHINS",
    "DAYWAIT",
    "DSMCRIT",
    "SUB2",
    "SUB3",
    "ALCFLG",
    "MARFLG",
    "COKEFLG",
    "HERFLG",
    "METHFLG",
    "BENZFLG",
    "OTHERFLG",
    "VET"
]

all_data=all_data.select(important_columns)
```
